```python
import os
import numpy as np
import pandas as pd

base_dir = "results"

rows = []

for root, dirs, files in os.walk(base_dir):
    for file in files:
        if file == "performance.npy":
            path = os.path.join(root, file)

            # Load performance object (object array containing dict)
            try:
                arr = np.load(path, allow_pickle=True)
                perf = arr.item() if isinstance(arr, np.ndarray) else arr
            except Exception as e:
                print(f"Failed to load {path}: {e}")
                continue

            # Initialize fields
            row = {
                "dataset": None,
                "shots": None,
                "model": None,
                "seed": None,
                "accuracy": None,
                "auc": None,
                "logloss": None,
                "train time": None,
                "test time": None,
            }

            # Parse metadata from path
            parts = root.split(os.sep)
            for part in parts:
                if part.startswith("seed="):
                    row["seed"] = int(part.split("=")[1])
                elif part.startswith("shot="):
                    row["shots"] = int(part.split("=")[1])
                elif part.startswith("model="):
                    row["model"] = part.split("=")[1]
                elif part.startswith("data="):
                    row["dataset"] = int(part.split("=")[1])

            # Extract performance metrics
            if isinstance(perf, dict):
                train = perf.get("Train", {})
                test = perf.get("Test", {})

                row["accuracy"] = test.get("accuracy", None)
                row["auc"] = test.get("auc", None)
                row["logloss"] = test.get("logloss", None)

                row["train time"] = perf.get("Average train time", None)
                row["test time"] = perf.get("Average test time", None)

            rows.append(row)

# Create DataFrame
df = pd.DataFrame(rows)

# Optional: reorder columns
df = df[
    ["dataset", "shots", "model", "seed",
     "accuracy", "auc", "logloss", "train time", "test time"]
]

df

# Optional: save
# df.to_csv("results_summary.csv", index=False)
```

```python
import os
import re
import ast
import pandas as pd
from collections import defaultdict

log_dir = "logs"

rows = []

for fname in os.listdir(log_dir):
    if "mlp" not in fname or not fname.endswith(".log"):
        continue

    path = os.path.join(log_dir, fname)

    # --- Extract metadata from filename ---
    match = re.match(r"id(\d+)_shot(\d+)_seed(\d+)_mlp\.log", fname)
    if not match:
        continue

    dataset, shots, seed = match.groups()

    best_params = None
    scores = None

    # --- Parse file ---
    with open(path, "r") as f:
        for line in f:
            if "Best params:" in line:
                best_params = ast.literal_eval(line.split("Best params:")[1].strip())

            if "test score:" in line:
                scores = ast.literal_eval(line.split("test score:")[1].strip())

    if best_params is None or scores is None:
        continue

    rows.append({
        "dataset": dataset,
        "shots": int(shots),
        "seed": int(seed),
        "params": best_params,
        "accuracy": scores.get("accuracy"),
        "auc": scores.get("auc"),
        "logloss": scores.get("logloss"),
    })

# Convert to DataFrame
df = pd.DataFrame(rows)

print("\n=== Raw Results ===")
print(df)

# ---------------------------------------------------
# 🔥 Aggregate: best params across datasets per shot
# ---------------------------------------------------

grouped = defaultdict(list)

for _, row in df.iterrows():
    key = (row["shots"], str(row["params"]))  # params as string for grouping
    grouped[key].append(row)

summary = []

for (shots, params_str), items in grouped.items():
    acc = sum(x["accuracy"] for x in items) / len(items)
    auc = sum(x["auc"] for x in items) / len(items)

    summary.append({
        "shots": shots,
        "params": params_str,
        "mean_accuracy": acc,
        "mean_auc": auc,
        "count": len(items)
    })

summary_df = pd.DataFrame(summary)

# Pick best params per shot (by accuracy)
best_per_shot = summary_df.sort_values("mean_accuracy", ascending=False)\
                         .groupby("shots")\
                         .first()\
                         .reset_index()

print("\n=== Best Params per Shot ===")
print(best_per_shot)
```
